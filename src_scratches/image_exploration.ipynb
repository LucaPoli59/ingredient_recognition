```python
import matplotlib.pyplot as plt
import os
import cv2
import random
import numpy as np

from config import *
from commons import *
```

```python
def read_image(img_file):
    img = cv2.imread(img_file, cv2.IMREAD_COLOR)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
seed = 20
random.seed(seed)

images_it = os.path.join(IMAGES_PATH, 'italian')
image_id = random.randint(0, len(os.listdir(images_it)))
image_file = os.path.join(images_it, os.listdir(images_it)[image_id])
```

```python
image = read_image(image_file)
show_image(image)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(image[:, :, 0], cmap='Reds')
axs[1].imshow(image[:, :, 1], cmap='Greens')
axs[2].imshow(image[:, :, 2], cmap='Blues')
for ax in axs:
    ax.axis('off')
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
show_image(image_gray, cmap='gray', title='Gray Image')
```

```python
# Find different shape image
image_diff_shape = read_image(os.path.join(IMAGES_PATH, 'italian', 'Pappardelle-Puttanesca-Allrecipes.jpg'))
show_image(image_diff_shape)
```

```python
image_resized = cv2.resize(image, None, fx=0.5, fy=0.5)
show_image(image_resized, title="Resized Image of 50%")
image_resized = cv2.resize(image, (100, 200))
show_image(image_resized, title="Resized Image to 100x200")
```

```python
image_diff_resized = cv2.resize(image_diff_shape, (image.shape[0], image.shape[1]), interpolation=cv2.INTER_CUBIC)
show_image(image_diff_resized)
```

```python
sharp_kernel = np.array([[-1, -1, -1],
                         [-1, 9, -1],
                         [-1, -1, -1]])
image_sharp = cv2.filter2D(image, -1, sharp_kernel)
show_image(image_sharp)
```

```python
blur_kernel = np.ones((3, 3), np.float32) / 10
image_blur = cv2.filter2D(image, -1, blur_kernel)
show_image(image_blur)
```
